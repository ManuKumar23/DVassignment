```python
import pandas as pd
import numpy as np

file_path = 'tiktok_performance.csv'
tiktok_data = pd.read_csv(file_path)
display(tiktok_data)

mean_likes = tiktok_data['Likes'].mean()
median_likes = tiktok_data['Likes'].median()
sum_likes = tiktok_data['Likes'].sum()
max_likes = tiktok_data['Likes'].max()

print("\nBasic Operations:")
print(f"Mean Likes: {mean_likes}")
print(f"Median Likes: {median_likes}")
print(f"Total Likes: {sum_likes}")
print(f"Max Likes: {max_likes}")


correlation_matrix = tiktok_data[['Likes', 'Comments', 'Shares', 'Views']].corr()
print("\nCorrelation Matrix:\n", correlation_matrix)

popular_videos = tiktok_data[(tiktok_data['Likes'] > 1000) & (tiktok_data['Views'] > 30000)]
print("\nFiltered Popular Videos:\n", popular_videos[['Video_ID', 'Likes', 'Views']])

print("\nAdvanced Operations:")
category_stats = tiktok_data.groupby('Category')[['Likes', 'Comments', 'Shares']].mean()
print("\nAverage Metrics by Category:\n", category_stats)

tiktok_data['Engagement_Rate'] = (tiktok_data['Likes'] + tiktok_data['Comments'] + tiktok_data['Shares']) / tiktok_data['Views']
print("\nEngagement Rate for each video:\n", tiktok_data[['Video_ID', 'Engagement_Rate']])
```
