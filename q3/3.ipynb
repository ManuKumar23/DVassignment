```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

tips = sns.load_dataset("tips")

plt.figure(figsize=(15, 10))

# Comparison Plot 1: Scatter
plt.subplot(2, 2, 1)
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title("Comparison Plot: Total Bill vs Tip")

# Comparison Plot 2: Box Plot to compare 'total_bill' across different days
plt.subplot(2, 2, 2)
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Comparison Plot: Total Bill Across Days by Gender")

# Composition Plot 1: Pie Chart to show percentage of male vs female
plt.subplot(2, 2, 3)
gender_counts = tips['sex'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99'])
plt.title("Composition Plot: Gender Distribution")

# Composition Plot 2: Stacked Bar Plot for smoker vs non-smoker over days
plt.subplot(2, 2, 4)
smoker_day = pd.crosstab(tips['day'], tips['smoker'])
smoker_day.plot(kind="bar", stacked=True, color=['#ff9999','#66b3ff'], ax=plt.gca())
plt.title("Composition Plot: Smoker vs Non-Smoker by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```
