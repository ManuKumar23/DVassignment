```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate some sample data
np.random.seed(42)
data = pd.DataFrame({
    'category': np.random.choice(['A', 'B', 'C'], size=100),
    'value': np.random.randn(100)
})

# Set the aesthetic style of the plots
sns.set(style="whitegrid")  # Set the style to whitegrid

# Create a boxplot with custom aesthetics
plt.figure(figsize=(10, 6))
sns.boxplot(x='category', y='value', data=data, palette='pastel', linewidth=1.5)

# Customize the aesthetics
plt.title('Boxplot of Values by Category', fontsize=16, fontweight='bold')
plt.xlabel('Category', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.grid(True)  # Add grid lines
plt.xticks(fontsize=12)  # Customize x-tick labels
plt.yticks(fontsize=12)  # Customize y-tick labels

# Show the plot
plt.show()

# Set another aesthetic context
sns.set_context("talk")  # Change context for larger font sizes
plt.figure(figsize=(10, 6))
sns.violinplot(x='category', y='value', data=data, palette='muted', inner='quartile')

# Customize the aesthetics
plt.title('Violin Plot of Values by Category', fontsize=16, fontweight='bold')
plt.xlabel('Category', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.grid(True)  # Add grid lines
plt.xticks(fontsize=12)  # Customize x-tick labels
plt.yticks(fontsize=12)  # Customize y-tick labels

# Show the plot
plt.show()
```
