```python
import pandas as pd
import numpy as np

file_path = 'AAPL.csv'
apple_data = pd.read_csv(file_path)
display(apple_data)
close_prices = apple_data['Close']

mean_pandas = close_prices.mean()
median_pandas = close_prices.median()
mode_pandas = close_prices.mode().iloc[0]  
std_dev_pandas = close_prices.std()


mean_numpy = np.mean(close_prices)
median_numpy = np.median(close_prices)
std_dev_numpy = np.std(close_prices, ddof=1)  

print("Using Pandas:")
print(f"Mean: {mean_pandas}")
print(f"Median: {median_pandas}")
print(f"Mode: {mode_pandas}")
print(f"Standard Deviation: {std_dev_pandas}")

print("\nUsing Numpy:")
print(f"Mean: {mean_numpy}")
print(f"Median: {median_numpy}")
print(f"Standard Deviation: {std_dev_numpy}")
```
