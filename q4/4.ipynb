```python
import numpy as np
import matplotlib.pyplot as plt

# Generating sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)  
y2 = np.cos(x)  
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

# 1. Line Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine', color='blue')
plt.plot(x, y2, label='Cosine', color='orange')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.show()

# 2. Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='green', alpha=0.5, label='Sine Points')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.show()

# 3. Bar Plot
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='purple')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

# 4. Histogram
data = np.random.randn(1000)  # Generate random data
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='cyan', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()

# 5. Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Pie Chart')
plt.axis('equal') 
plt.show()

# 6. Box Plot
plt.figure(figsize=(10, 6))
data_for_box = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data_for_box, vert=True, patch_artist=True, labels=['1', '2', '3'])
plt.title('Box Plot')
plt.xlabel('Distribution')
plt.ylabel('Values')
plt.grid()
plt.show()

# 7. Area Plot
plt.figure(figsize=(10, 6))
plt.fill_between(x, y, color='skyblue', alpha=0.5, label='Sine Area')
plt.fill_between(x, y2, color='orange', alpha=0.5, label='Cosine Area')
plt.title('Area Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.show()
```
