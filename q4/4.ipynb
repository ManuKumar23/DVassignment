```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'crop_yield.csv'
crop_yield_data = pd.read_csv(file_path)
display(crop_yield_data)
# Calculate the average yield per year
avg_yield_per_year = crop_yield_data.groupby('Crop_Year')['Yield'].mean()

# Plotting the average yield per year
plt.figure(figsize=(10, 6))
plt.plot(avg_yield_per_year.index, avg_yield_per_year.values, marker='o', linestyle='-', color='b', label='Average Yield')

# Adding plot details
plt.title('Average Crop Yield Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Yield (metric tons per hectare)')
plt.legend(loc='upper right')
plt.grid(True)
plt.xticks(rotation=45)

# Adding some basic text to indicate notable years
plt.text(2000, avg_yield_per_year.loc[2000], 'Year 2000', ha='center', color='red')
plt.text(2010, avg_yield_per_year.loc[2010], 'Year 2010', ha='center', color='green')

# Show the plot
plt.tight_layout()
plt.show()
```
